# Ejercicio Bootstrap.

```
install.packages('boot')
```

```
# Poblacion de estudio

library(MASS)
mu=c(2.5,4)
Sigma=matrix(c(4,4.5,4.5,9),2,2) 
X<-mvrnorm(n=200,mu,Sigma)
#print(X)
```

```
#parámetros sobre los que daremos IC
mean(X[,1])
mean(X[,2])
var(X)
```

```
library(boot)

data_mean <- function(data, i){
  sample <- data[i,]
  
  c(mean(sample[,1]), mean(sample[,2]),  var(sample))
}
bootstrap <- boot(X, data_mean, R=50)
bootstrap
```

```
# primer estimador: media

# media 1
boot.ci(bootstrap, type="perc",index=c(1))

# media 2
boot.ci(bootstrap, type="perc",index=c(2))  
```

```
# segundo estimador: matriz var-cov

# var(1)
boot.ci(bootstrap, type="perc",index=c(3)) 

# cov(1,2) = cov(2,1)
boot.ci(bootstrap, type="perc",index=c(4)) 

# var(2)
boot.ci(bootstrap, type="perc",index=c(6)) 
```
